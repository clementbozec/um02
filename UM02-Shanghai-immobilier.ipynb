```python
# Seaborn for plotting and styling
import seaborn as sns
```

```python
# Boxplot
sns.boxplot(data=df)
plt.yscale("log")
```

```python
# Pre-format DataFrame
stats_df = df.drop(['done', 'longitude', 'latitude'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
plt.yscale("log")
```

```python
# Set theme
sns.set_style('whitegrid')
 
# Violin plot
plt.figure(figsize=(15,10))
sns.violinplot(x='district', y='price', data=df)
plt.yscale("log")
```

```python
district_type_colors = ['#78C850',  # baoshan
                    '#F08030',  # chongming
                    '#6890F0',  # fengxian
                    '#A8B820',  # jiading
                    '#A8A878',  # jingan
                    '#A040A0',  # jinshan
                    '#F8D030',  # minhang
                    '#E0C068',  # pudong
                    '#EE99AC',  # qingpu
                    '#C03028',  # songjiang district
                    '#F85888',  # yangpu
                    '#B8A038',  # zhabei
                    '#705898',  # huangpu
                    '#98D8D8',  # xuhui
                    '#7038F8',  # putuo
                   ]
```

```python
# Violin plot with district color palette
plt.figure(figsize=(15,10))
sns.violinplot(x='district', y='price', data=df, 
               palette=district_type_colors) # Set color palette
plt.yscale("log")
```

```python
import numpy as np
```

```python
df['abc'] = np.random.uniform(0, 1, len(df)) <= 0.001
df = df[df['abc'] == True]
len(df)
```

```python
# Swarm plot with district color palette
plt.figure(figsize=(15,10))
sns.swarmplot(x='district', y='price', data=df, 
              palette=district_type_colors)
plt.yscale("log")
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(15,10))
 
# Create plot
sns.violinplot(x='district',
               y='price', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=district_type_colors)
plt.yscale("log")
 
sns.swarmplot(x='district', 
              y='price', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Price by district')
```

```python
stats_df = df.drop(['tags', 'done', 'abc'], axis=1)
```

```python
stats_df.head()
```

```python
# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["surface", "address", "district", "neighborhood", "price", "onesquaremeter", "longitude", "latitude"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```

```python
print( stats_df.shape )
print( melted_df.shape )
# (151, 9)
# (906, 5)
```

```python
# Swarmplot with melted_df
sns.swarmplot(x='Stat', y='value', data=melted_df, 
              hue='district')
```

```python
# 1. Enlarge the plot
plt.figure(figsize=(15,10))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='district', 
              dodge=True, # 2. Separate points by hue
              palette=district_type_colors) # 3. Use Pokemon palette
 
# 4. Adjust the y-axis
plt.ylim(0, 7)
 
# 5. Place legend to the right
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
melted_df['abcd'] = np.random.uniform(0, 1, len(melted_df)) <= 0.1
melted_df = melted_df[melted_df['abcd'] == True]
len(melted_df)
```

```python
# Calculate correlations
corr = stats_df.corr()
 
# Heatmap
plt.figure(figsize=(15,10))
sns.heatmap(corr)
```

```python
# Distribution Plot (a.k.a. Histogram)
plt.figure(figsize=(15,10))
sns.distplot(df.surface)
```
